```python
import matplotlib.pyplot as plt
import matplotlib.path as mplPath
import numpy as np
import pandas as pd

%matplotlib inline
```

## Import data

```python
with open("master1.txt","r") as f:
    master1 = f.readlines()
list1 = [line.strip('\n') for line in master1]

list2 = []
for x in range(len(list1)):
    line = list1[x].split('  ')
    line = [string for string in line if ((string != '') & (string != ' '))]
    for index in range(len(line)):
        if line[index][0] == ' ':
            line[index] = line[index][1:]
        if index in (2,3,7):
            line[index] = float(line[index])
            if ((line[index] > 0) & (str(line[index])[::1].find('.') == 1)):
                line[index] = float(line[index])*10
            elif ((line[index] < 0) & (str(line[index])[::1].find('.') == 2)):
                line[index] = float(line[index])*10                    
        elif index in (0,4,5,6):
            line[index] = int(line[index])
    if len(line) == 8:
        line.append('UNKNOWN')
    list2.append(line)

for i in range(1,len(list2)):
    if (list2[i][7] < list2[i-1][7]) & (list2[i][0] == list2[i-1][0]):
        list2[i][7] = list2[i-1][7] + 0.25
```

```python
columns = ['ID', 'name', 'lat', 'long', 'ws', 'cp', 'year', 'day', 'cat']
df = pd.DataFrame(list2, columns = columns)
```

```python
df.to_csv('master2.csv')
```

## 2. Frequency of events

```python
hurr_years = np.array([df[df['ID'] == item].iloc[0, 6] for item in hurr_ID])
hurr_dec = np.ndarray.astype((np.floor(hurr_years/10)*10), int)

decades, dec_counts = np.unique(hurr_dec, return_counts=True)
```

```python
def least_sq_line(x1,y1):
    N = len(x1)
    x2 = [x**2 for x in x1]
    y2 = [y**2 for y in y1]
    xy = [x1[i] * y1[i] for i in range(N)]
    
    delta = (N * np.sum(x2)) - (np.sum(x1))**2
    A = (1/delta) * ( np.multiply(np.sum(x2), np.sum(y1)) - np.multiply(np.sum(x1), np.sum(xy)) )
    B = (1/delta) * ( (N * np.sum(xy)) - np.multiply(np.sum(x1), np.sum(y1)) )
    
    y_err = np.sqrt(np.sum([(y1[i] - A - (B*x1[i]))**2 for i in range(N)]) / (N-2))
    
    return A, B, y_err
```

```python
# Linear fit
A, B, y_err = least_sq_line(decades, dec_counts)

# Extend fit to 2030
x = np.append(decades, [2010, 2020, 2030])
y_lower = [x[i]*B + A - y_err for i in range(len(x))]
y_upper = [x[i]*B + A + y_err for i in range(len(x))]

plt.figure(figsize=(10,6))
plt.plot(decades, dec_counts, ".-", markersize=16, label="Events from data")
plt.plot(x, A + B*x, 'r-', label='Least-squares fit')
plt.fill_between(x, y_lower, y_upper, alpha=.5)
plt.xticks(x, size=12)
plt.yticks(np.arange(0, np.max(dec_counts)*1.1,10), size=12)
plt.title("Frequency of hurricane events by decade", size=20)
plt.xlabel("Decade (by start year)", size=16)
plt.ylabel("Hurricanes per decade", size=16)
for i in x[::2]:
    plt.axvspan(i, i+10, facecolor="grey", alpha=0.05)
ax = plt.axes()
ax.xaxis.grid(alpha=0.5)

plt.legend(loc="lower right", fontsize=14)
plt.show()
```

## Florida impact

```python
# Coordinates of Florida polygon
vertices = [(25.1, -79.3),
      (30.8, -81.5),
      (30.8, -87.6),
      (30.0, -87.6),
      (30.0, -83.8),
      (25.1, -81.7),
      (25.1, -79.3)]

codes = [mplPath.Path.MOVETO,
        mplPath.Path.LINETO,
        mplPath.Path.LINETO,
        mplPath.Path.LINETO,
        mplPath.Path.LINETO,
        mplPath.Path.LINETO,
        mplPath.Path.LINETO]

# Create Path object for Florida
fl_path = mplPath.Path(vertices, codes)
```

```python
impact = pd.DataFrame(columns = columns)

for i in range(len(hurr_ID)):
    df_i = hurr_data[hurr_ID[i]]
    for j in range(len(df_i)):
        index = df_i.index[j]
        if fl_path.contains_point(tuple(df_i.loc[index, ['lat', 'long']])):
            impact.loc[i,:] = df_i.loc[index,:]
            t_start = df_i.loc[index, 'day']
            t_stop = t_start
            for k in range(j+1, len(df_i)):
                index_k = df_i.index[k]
                if fl_path.contains_point(tuple(df_i.loc[index_k, ['lat', 'long']])):
                    t_stop = df_i.loc[index_k, 'day']
                else:
                    break
            impact.loc[i, 'dt'] = (t_stop - t_start) * 24
            break
```

```python
x = np.array(impact.loc[:,'year'])
y = np.array(impact.loc[:,'ws']) * np.array(impact.loc[:,'dt'])
x1 = x[y > 0]
y1 = y[y > 0]

plt.figure(figsize=(10,6))
plt.plot(x, y, ".", markersize=16)
plt.xticks(np.arange(1900, 2020, 10), size=12)
plt.yticks(np.arange(0,np.max(y)*1.1,500), size=12)
plt.title("Impact factor = (impact windspeed) $\\times$ (duration)", size=20)
plt.xlabel("Year", size=16)
plt.ylabel("Impact factor", size=16)
plt.grid('on')

plt.show()

plt.figure(figsize=(10,6))
plt.plot(x1, y1, ".", markersize=16)
plt.xticks(np.arange(1900, 2020, 10), size=12)
plt.yticks(np.arange(0,np.max(y)*1.1,500), size=12)
plt.title("Impact factor = (impact windspeed) $\\times$ (duration)", size=20)
plt.xlabel("Year", size=16)
plt.ylabel("Impact factor", size=16)
plt.grid('on')

plt.show()
```
