```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
cols = ['year'] + months + ['annual']
cetMaxRaw = pd.DataFrame(np.genfromtxt('cetMax.txt'), columns=cols)
cetMinRaw = pd.DataFrame(np.genfromtxt('cetMin.txt'), columns=cols)
```

```python
years = np.arange(1878,2018,1/12)
maxT = np.array([[x for x in np.array(row[months])] for i, row in cetMaxRaw.iterrows()]).flatten()
minT = np.array([[x for x in np.array(row[months])] for i, row in cetMinRaw.iterrows()]).flatten()
cetMax = pd.DataFrame({'year':years, 'monthly':maxT})
cetMin = pd.DataFrame({'year':years, 'monthly':minT})
cetMax = cetMax[cetMax['monthly']>-99]
cetMin = cetMin[cetMin['monthly']>-99]

years = cetMax['year']
maxT = cetMax['monthly']
minT = cetMin['monthly']
```

```python
maxBoxcarAvg = [np.mean(maxT[i-30:i+31]) for i in range(30,len(years)-30)]
minBoxcarAvg = [np.mean(minT[i-30:i+31]) for i in range(30,len(years)-30)]
```

```python
gaussian = lambda x, mu, sig: np.exp(-(x-mu)**2/(2*sig**2))
normGaussian = lambda x, mu, sig: gaussian(x,mu,sig)/(np.sum(gaussian(x,mu,sig)))

maxGaussianAvg = [np.sum(maxT*normGaussian(years, i, 7*12)) for i in years]
minGaussianAvg = [np.sum(minT*normGaussian(years, i, 7*12)) for i in years]
```

```python
def expSmooth(x,y,alpha):
    ySmooth = [y[0]]
    for i in range(1,len(x)):
        ySmooth.append(alpha*y[i] + (1-alpha)*y[i-1])
    return ySmooth
```

```python
maxExpAvg = expSmooth(years,maxT,.5)
minExpAvg = expSmooth(years,minT,.5)
```

```python
plt.figure(figsize=(12,5))
plt.plot(years, maxT, '.', label='Unsmoothed')
plt.plot(years[30:-30], maxBoxcarAvg, 'k-', label='Boxcar')
plt.plot(years, maxGaussianAvg, 'g-', label='Gaussian')
plt.plot(years, maxExpAvg, 'r-', label='Exponential')

plt.xticks(np.arange(1880,2021,10), size=14)
plt.yticks(size=14)
plt.title('Maximum temperatures smoothing', size=20)
plt.xlabel('Year', size=16)
plt.ylabel('T ($^\circ$C)', size=16)

plt.legend(loc='lower right', fontsize=14)
plt.grid()

plt.savefig('smoothingMaxT.png')
plt.show()
```
